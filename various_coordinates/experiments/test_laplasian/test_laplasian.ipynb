```python
def norm_plot(model, grid):
    plt.figure(figsize=(10, 3))
    plt.title('Непрерывная норма градиента ({})'.format(grid))
    plt.xlabel('время')
    plt.yscale('log')
    plt.plot(model.ts, model.df_additional.iloc[:, 0])
    plt.show()
```
