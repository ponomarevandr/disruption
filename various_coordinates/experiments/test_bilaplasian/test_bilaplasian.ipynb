## Прямоугольные координаты

```python
model_50 = model_from_file('../../bin/model_50.csv')
model_plot(model_50, [0, 0.5, 1, 7])
```

```python
def norm_plot(model, grid):
    plt.figure(figsize=(10, 3))
    plt.title('Непрерывная норма градиента ({})'.format(grid))
    plt.xlabel('время')
    plt.yscale('log')
    plt.plot(model.ts, model.df_additional.iloc[:, 0])
    plt.show()
```

```python
norm_plot(model_50, '$n_x = 50$')
```

```python
model_100 = model_from_file('../../bin/model_100.csv')
model_plot(model_100, [0, 0.5, 1, 7])
```

```python
norm_plot(model_100, '$n_x = 100$')
```

```python
model_200 = model_from_file('../../bin/model_200.csv')
norm_plot(model_200, '$n_x = 200$')
```

```python
plt.figure(figsize=(10, 7))
plt.title('Стационарные решения в прямоугольных координатах')
plt.xlim([-0.05, 3])
plt.xlabel('$r$')
plt.plot([0, 0], [0, 1], '--', color='black', label='_nolegend_')
plt.plot([0, 5], [1, 1], '--', color='black', label='_nolegend_')
plt.plot(model_50.xs, model_50.phi_at_t(7))
plt.plot(model_100.xs, model_100.phi_at_t(7))
plt.plot(model_200.xs, model_200.phi_at_t(7))
plt.legend(['$n_x = 50$', '$n_x = 100$', '$n_x = 200$'])
plt.show()
```
