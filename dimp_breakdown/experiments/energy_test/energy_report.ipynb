```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data_dir = 'data/'
```

```python
plt.rcParams.update({'font.size': 12})
figsize=(7, 5.5)
```

```python
df_disturbance_adapt = pd.read_csv(
    data_dir + 'disturbance_new_adaptive.dat',
    sep=r'\s+', index_col=False
)
df_disturbance_adapt
```

```python
df_disturbance_energy = pd.read_csv(
    data_dir + 'disturbance_new_energy.dat',
    sep=r'\s+', index_col=False
)
df_disturbance_energy['energy_total'] = df_disturbance_energy['energy_electr'] + df_disturbance_energy['energy_nonlocal']
df_disturbance_energy
```

```python
plt.figure(figsize=figsize, dpi=300)
plt.xlabel('временной слой')
plt.plot(df_disturbance_adapt.index, df_disturbance_adapt['phi_mean'], color='C0')
plt.legend([r'среднее фазового поля $\phi$'])
plt.show()
```

```python
plt.figure(figsize=figsize, dpi=300)
plt.xlabel('временной слой')
plt.plot(df_disturbance_energy.index, df_disturbance_energy['energy_total'], color='C1')
plt.legend([r'свободная энергия $\Psi$'], loc='lower left')
plt.show()
```

```python
df_needle_adapt = pd.read_csv(
    data_dir + 'needle_new_adaptive.dat',
    sep=r'\s+', index_col=False
)
df_needle_adapt
```

```python
df_needle_energy = pd.read_csv(
    data_dir + 'needle_new_energy.dat',
    sep=r'\s+', index_col=False
)
df_needle_energy['energy_total'] = df_needle_energy['energy_electr'] + df_needle_energy['energy_nonlocal']
df_needle_energy
```

```python
plt.figure(figsize=figsize, dpi=300)
plt.xlabel('временной слой')
plt.plot(df_needle_adapt.index, df_needle_adapt['phi_mean'], color='C0')
plt.legend([r'среднее фазового поля $\phi$'])
plt.show()
```

```python
plt.figure(figsize=figsize, dpi=300)
plt.xlim([30000, 51500])
plt.xticks(np.linspace(30000, 50000, 5))
plt.xlabel('временной слой')
plt.plot(df_needle_energy.index, df_needle_energy['energy_total'], color='C1')
plt.legend([r'свободная энергия $\Psi$'], loc='lower left')
plt.show()
```
