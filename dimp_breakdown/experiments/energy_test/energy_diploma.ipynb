```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data_dir = 'data/'
```

```python
df_disturbance_adapt = pd.read_csv(
    data_dir + 'disturbance_new_adaptive.dat',
    sep=r'\s+', index_col=False
)
df_disturbance_adapt
```

```python
df_disturbance_energy = pd.read_csv(
    data_dir + 'disturbance_new_energy.dat',
    sep=r'\s+', index_col=False
)
df_disturbance_energy['energy_total'] = df_disturbance_energy['energy_electr'] + df_disturbance_energy['energy_nonlocal']
df_disturbance_energy
```

```python
disturbance_views = np.array([1000, 4000, 7000])
```

```python
plt.rcParams.update({'font.size': 12})
figsize=(10, 3)
```

```python
plt.figure(figsize=figsize, dpi=300)
plt.grid(True)
plt.xlabel('временной слой')
plt.ylabel(r'$\phi$')
plt.plot(df_disturbance_adapt.index, df_disturbance_adapt['phi_mean'], color='black') #='C0')
plt.scatter(disturbance_views, df_disturbance_adapt['phi_mean'][disturbance_views], marker='x', s=100, color='black')
plt.legend([r'среднее фазового поля $\phi$', 'моменты, изображенные на рисунках'])
plt.show()
```

```python
plt.figure(figsize=figsize, dpi=300)
plt.grid(True)
plt.xlabel('временной слой')
plt.ylabel(r'$\Psi$, Дж')
plt.plot(df_disturbance_energy.index, df_disturbance_energy['energy_total'], color='black') #='C1')
plt.scatter(disturbance_views, df_disturbance_energy['energy_total'][disturbance_views], marker='x', s=100, color='black')
plt.legend([r'свободная энергия $\Psi$', 'моменты, изображенные на рисунках'], loc='lower left')
plt.show()
```
