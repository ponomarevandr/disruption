```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
calcs_dir = '/home/andrey/Desktop/kiam/calculations_k60/'
```

```python
df_disturbance_adapt = pd.read_csv(
    calcs_dir + 'test_129_disturbance_26-03-16/OUTPUT/adaptive_time_step.dat',
    sep=r'\s+', index_col=False
)
df_disturbance_adapt
```

```python
df_disturbance_energy = pd.read_csv(
    calcs_dir + 'test_129_disturbance_26-03-16/OUTPUT/energy.dat',
    sep=r'\s+', index_col=False
)
df_disturbance_energy['energy_total'] = df_disturbance_energy['energy_electr'] + df_disturbance_energy['energy_nonlocal']
df_disturbance_energy
```

```python
plt.figure(figsize=(9, 5))
plt.plot(df_disturbance_adapt['model_time'], df_disturbance_adapt['phi_mean'])
plt.plot(df_disturbance_adapt['model_time'][::100], df_disturbance_adapt['phi_mean'][::100], 'x', color='black')
plt.show()
```

```python
plt.figure(figsize=(9, 5))
plt.plot(df_disturbance_adapt.index, df_disturbance_adapt['phi_mean'])
plt.show()
```

```python
plt.figure(figsize=(9, 5))
plt.plot(df_disturbance_adapt['model_time'], df_disturbance_energy['energy_total'], color='C1')
plt.plot(df_disturbance_adapt['model_time'][::100], df_disturbance_energy['energy_total'][::100], 'x', color='black')
plt.show()
```

```python
plt.figure(figsize=(9, 5))
plt.plot(df_disturbance_adapt.index, df_disturbance_energy['energy_total'], color='C1')
plt.show()
```

```python
plt.figure(figsize=(9, 5))
plt.yscale('log')
plt.plot(df_disturbance_adapt.index, -df_disturbance_energy['energy_electr'], color='C3')
plt.plot(df_disturbance_adapt.index, df_disturbance_energy['energy_nonlocal'], color='C2')
plt.show()
```

```python
plt.figure(figsize=(9, 7))
plt.xlim([2800, 3200])
plt.ylim([-4e-9, -1e-9])
plt.plot(df_disturbance_adapt.index, df_disturbance_energy['energy_total'], color='C1')
plt.show()
```

```python
plt.figure(figsize=(9, 7))
plt.xlim([2950, 3075])
plt.ylim([-2.5e-9, -1.8e-9])
plt.plot(df_disturbance_adapt.index, df_disturbance_energy['energy_total'], '-o', color='C1')
plt.show()
```
