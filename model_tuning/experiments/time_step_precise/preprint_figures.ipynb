```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.rcParams.update({'font.size': 12})
```

```python
delta = 0.001

def f(phi):
    return 4 * phi**3 - 3 * phi**4

def f_phi(phi):
    return 12 * phi**2 - 12 * phi**3
    
def f_phi_phi(phi):
    return 24 * phi - 36 * phi**2

def eps(phi):
    return 1 / (f(phi) + delta)

def eps_phi(phi):
    return -1 * f_phi(phi) / (f(phi) + delta)**2

def eps_phi_phi(phi):
    return (
        2 * f_phi(phi)**2 - f_phi_phi(phi) * (f(phi) + delta)
    ) / (
        (f(phi) + delta)**3
    )

def G(phi):
    return 12 * phi * (
        16 * phi**3 - 30 * phi**4 + 15 * phi**5 + 3 * delta * phi + 2 * delta
    ) / (
        (4 * phi**3 - 3 * phi**4 + delta)**3
    )
```

```python
phis = np.linspace(0, 1, 10000)
```

```python
plt.figure(figsize=(7.5, 4.5), dpi=300)
plt.grid(True)
plt.xlabel(r'$\phi$')
plt.xlim([-0.01, 0.25])
plt.yticks(np.arange(4) * 2 * 10**5)
plt.ticklabel_format(axis='y', style='sci', scilimits=(0,0))
plt.plot(phis, np.abs(eps_phi_phi(phis)), '--', color='black')
plt.plot(phis, np.abs(G(phis)), color='black')
plt.legend([r"$|\epsilon''(\phi) / \epsilon_0|$", r'$G(\phi)$'])
plt.show()
```
