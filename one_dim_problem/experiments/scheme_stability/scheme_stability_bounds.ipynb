```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams.update({'font.size': 14})

%matplotlib inline
```

```python
def f(phi):
    return 4 * phi ** 3 - 3 * phi ** 4

def f_phi(phi):
    return 12 * phi ** 2 - 12 * phi ** 3

def f_phi_phi(phi):
    return 24 * phi - 36 * phi ** 2

def eps_phi_phi(delta, phi):
    return (2 * f_phi(phi) ** 2 - (f(phi) + delta) * f_phi_phi(phi)) / (f(phi) + delta) ** 3
```

```python
phi = np.linspace(0, 0.3, 700)
plt.figure(figsize=(8, 4))
plt.grid(True)
plt.xlabel('$\phi$')
plt.yticks([-400000, -200000, 0, 200000])
plt.plot(phi, eps_phi_phi(0.001, phi))
plt.legend(['$\epsilon\'\'(\phi), \; \delta = 0.001$'])
plt.show()
```

```python
phi = np.linspace(0, 0.5, 700)
plt.figure(figsize=(9, 5))
plt.grid(True)
plt.xlabel('$\phi$')
plt.plot(phi, 0.1 ** (5/3) * eps_phi_phi(0.1, phi), color='C2')
plt.plot(phi, 0.001 ** (5/3) * eps_phi_phi(0.001, phi), color='C0')
plt.plot(phi, 0.00001 ** (5/3) * eps_phi_phi(0.00001, phi), color='C1')
plt.plot([0, 0.5], 2 * [2.2161], '--', color='black')
plt.plot([0, 0.5], 2 * [-4.3782], '--', color='black')
plt.legend([
    '$\delta^{5/3} \epsilon\'\'(\phi), \; \delta = 10^{-1}$',
    '$\delta^{5/3} \epsilon\'\'(\phi), \; \delta = 10^{-3}$',
    '$\delta^{5/3} \epsilon\'\'(\phi), \; \delta = 10^{-5}$',
    'оценки экстремумов'
], bbox_to_anchor=(0.52, 0.51), loc="upper left")
plt.show()
```
