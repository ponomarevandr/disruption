```python
def model_plot_scaled(model, times, print_params=True, **kwargs):
    if 'figsize' in kwargs:
        plt.figure(figsize=kwargs['figsize'])
    if 'title' in kwargs:
        plt.title(kwargs['title'])
    if 'xlim' in kwargs:
        plt.xlim(kwargs['xlim'])
    if 'ylim' in kwargs:
        plt.ylim(kwargs['ylim'])
    plt.xlabel('$r$')
    plt.ylabel('$\phi$')
    t_indices = [model.t_index(t) for t in times]
    for i in t_indices:
        plt.plot(np.sqrt(2 * model.xs), model.phi[i])
    plt.legend(['$t = {:.2f}$'.format(i * model.params['dt_data']) for i in t_indices],
        bbox_to_anchor=(1.02, 1), loc="upper left")
    plt.figtext(0, 0, (
            'dt = {:.6e}\n' +
            'dx = {:.6e}\n' +
            'duration = {:.6e}\n' +
            'width = {:.6e}'
        ).format(
            model.params['dt'],
            model.params['dx'],
            model.params['duration'],
            model.params['width']
        ),
        va='top'
    )
    plt.show()
```
