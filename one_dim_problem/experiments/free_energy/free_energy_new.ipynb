```python
def energy_density_plot(model, times, colors):
    plt.figure(figsize=(8, 5))
    plt.xlabel('$x$')
    plt.ylabel('плотность энергии')
    t_indices = [model.t_index(t) for t in times]
    for i, color in zip(t_indices, colors):
        plt.plot(model.xs, model.phi[i], color='C{}'.format(color))
    plt.legend(['$t = {:.2f}$'.format(i * model.params['dt_data']) for i in t_indices],
        bbox_to_anchor=(1.02, 1), loc="upper left")
    plt.show()
```

```python
model_density_electrical = model_from_file('data/model_density_electrical.csv')
energy_density_plot(model_density_electrical,
                    [0, 0.5, 0.55, 0.6, 0.7, 0.8, 1],
                    [0, 4, 5, 6, 7, 8, 9])
```

```python
def energy_plot(model, times):
    ts = np.arange(model.df_phi.shape[0]) * model.params['dt_data']
    plt.figure(figsize=(10, 6))
    plt.xlabel('$t$')
    plt.ylabel('энергия')
    for t in times:
        if 0.05 < t < 0.25:
            plt.plot([t, t], [-55, 0], '--')
        else:
            plt.plot([t, t], [-275, 0], '--')
    plt.plot(ts, model.phi[:, 0], color='black')
    plt.legend(['$t = {:.2f}$'.format(t) for t in times],
               bbox_to_anchor=(0.05, 0.77), loc="upper left")
    plt.show()
```

```python
model_total = model_from_file('data/model_total.csv')
energy_plot(model_total, [0, 0.1, 0.3, 0.4, 0.5, 0.55, 0.6, 0.7, 0.8, 1])
```
