```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams.update({'font.size': 18})

%matplotlib inline
```

```python
def f(phi):
    return 4 * phi**3 - 3 * phi**4

def eps(phi, eps_0, delta):
    return eps_0 / (f(phi) + delta)

phi = np.linspace(0, 1, 600)
```

```python
plt.figure(figsize=(5, 5))
plt.xlabel('$\phi$')
plt.xticks([0, 1], [0, 1])
plt.yticks([0, 0.25, 0.5, 0.75, 1], [0, 0.25, 0.5, 0.75, 1])
plt.plot([0, 1], [0, 0], '--', color='black', label='_nolegend_')
plt.plot([0, 1], [1, 1], '--', color='black', label='_nolegend_')
plt.plot(phi, f(phi))
plt.legend(['$f(\phi)$'])
plt.show()
```

```python
higher = 5
lower = 0.2
eps_0 = 1 / (1 / lower - 1 / higher)
delta = 1 / (higher / lower - 1)
plt.figure(figsize=(5, 5))
plt.xlabel('$\phi$')
plt.xticks([0, 1], [0, 1])
plt.yticks([0], [0])
plt.plot([0, 1], [0, 0], '--', color='black', label='_nolegend_')
plt.plot([0, 1], [lower, lower], '--', color='C1', label='_nolegend_')
plt.plot([0, 1], [higher, higher], '--', color='C1', label='_nolegend_')
plt.plot(phi, eps(phi, eps_0, delta))
plt.legend(['$\epsilon(\phi)$'])
plt.show()
```
