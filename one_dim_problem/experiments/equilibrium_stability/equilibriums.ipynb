```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams.update({'font.size': 16})

%matplotlib inline
```

```python
phi = np.linspace(0, 1, 700)
```

```python
plt.figure(figsize=(8, 4))
plt.grid(True)
plt.xticks([0, 1])
plt.yticks(ticks=[0])
plt.plot([0, 1], [0, 0], '--', color='black', label='_nolegend_')
plt.plot(phi, phi**3 * (1 - phi), color='C0')
plt.scatter([0, 1], [0, 0], color='C0')
plt.legend(['$\chi(\phi)$'])
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.grid(True)
plt.xticks([0, 1])
plt.yticks(ticks=[0])
plt.plot([0, 1], [0, 0], '--', color='black', label='_nolegend_')
plt.plot(phi, -phi**4 * (phi - 0.6) * (phi - 1) * (1.5 - phi)**3, color='C0')
plt.scatter([0, 0.6, 1], [0, 0, 0], color='C0')
plt.legend(['$\chi(\phi)$'])
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.grid(True)
plt.xticks([0, 1])
plt.yticks(ticks=[0])
plt.plot([0, 1], [0, 0], '--', color='black', label='_nolegend_')
plt.plot(phi, phi**3 * (phi - 1), color='C0')
plt.scatter([0, 1], [0, 0], color='C0')
plt.legend(['$\chi(\phi)$'])
plt.show()
```
