```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def f(phi):
    return 4 * phi ** 3 - 3 * phi ** 4

def f_phi(phi):
    return 12 * phi ** 2 - 12 * phi ** 3

def f_phi_phi(phi):
    return 24 * phi - 36 * phi ** 2

def eps_phi_phi(delta, phi):
    return (2 * f_phi(phi) ** 2 - (f(phi) + delta) * f_phi_phi(phi)) / (f(phi) + delta) ** 3
```

```python
phi = np.linspace(0, 0.3, 500)
plt.figure(figsize=(7, 7))
plt.grid(True)
plt.xlabel('$\phi$')
plt.plot(phi, eps_phi_phi(0.001, phi))
plt.legend(['$\epsilon\'\'(\phi), \; \delta = 0.001$'])
plt.show()
```

```python
phi = np.linspace(0, 0.5, 500)
plt.figure(figsize=(7, 7))
plt.grid(True)
plt.xlabel('$\phi$')
plt.plot(phi, 0.1 ** (5/3) * eps_phi_phi(0.1, phi), color='C2')
plt.plot(phi, 0.001 ** (5/3) * eps_phi_phi(0.001, phi), color='C0')
plt.plot(phi, 0.00001 ** (5/3) * eps_phi_phi(0.00001, phi), color='C1')
plt.plot([0, 0.5], 2 * [2.2161], '--', color='black')
plt.plot([0, 0.5], 2 * [-4.3782], '--', color='black')
plt.legend([
    '$\epsilon\'\'(\phi), \; \delta = 0.1$',
    '$\epsilon\'\'(\phi), \; \delta = 0.001$',
    '$\epsilon\'\'(\phi), \; \delta = 0.00001$'
])
plt.show()
```

```python
def eps_phi(delta, phi):
    result = np.zeros(phi.shape)
    result[phi < 0] = -12.0 / delta ** 2 * phi[phi < 0] ** 2
    result[phi >= 0] = -f_phi(phi[phi >= 0]) / (f(phi[phi >= 0]) + delta) ** 2
    return result
```

```python
phi = np.linspace(-0.03, 0.3, 500)
plt.figure(figsize=(7, 7))
plt.grid(True)
plt.xlabel('$\phi$')
plt.plot(phi, eps_phi(0.001, phi))
plt.legend(['$\epsilon\'(\phi), \; \delta = 0.001$'])
plt.show()
```
