```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
dx = 10.0 / 1000.0
dt = 1.0 / 100.0
phi = pd.read_csv('../bin/phi_values.csv', sep=';', header=None)
```

```python
t_indices = [0, 20, 40, 60, 65, 67, 68, 70, 75, 80, 100]
plt.figure(figsize=(8, 6))
plt.xlabel('x')
plt.ylabel('phi')
for i in t_indices:
    values = phi.loc[i]
    plt.plot(np.arange(len(values)) * dx, values)
plt.legend([f't = {i * dt:.2f}' for i in t_indices])
plt.figtext(0, 0,
            'eps_0 = 1e-3\n' +
            'delta = 1e-6\n' +
            'l = 1.0\n' +
            'Gamma = 1.0\n' +
            'm = 1.0\n' +
            'Phi_coefficient = 60.0',
            va='top'
           )
plt.show()
```
