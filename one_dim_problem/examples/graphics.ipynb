```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
dx = 10.0 / 1000.0
dt = 1.0 / 100.0
phi = pd.read_csv('../bin/phi_values.csv', sep=';', header=None)
```

```python
t_indices = [0, 20, 40, 60, 65, 67, 68, 70, 75, 80, 100]
plt.figure(figsize=(8, 6))
for i in t_indices:
    values = phi.loc[i]
    plt.plot(np.arange(len(values)) * dx, values)
plt.legend([f't = {i * dt}' for i in t_indices])
plt.show()
```
