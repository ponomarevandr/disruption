```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

from math import sqrt

%matplotlib inline
```

```python
matplotlib.rcParams.update({'font.size': 14})
```

```python
xs = np.linspace(0, 1, 600)
```

```python
plt.figure(figsize=(3, 3)) #, dpi=300)
plt.grid(True)
plt.xticks([0, 1])
plt.yticks([0])
plt.plot(xs, xs**3 * (1 - xs), color='black')
plt.show()
```

```python
plt.figure(figsize=(3, 3)) #, dpi=300)
plt.grid(True)
plt.xticks([0, 1])
plt.yticks([0])
plt.plot(xs, -xs**4 * (1 - xs) * (0.62 - xs) * np.exp(-3 * xs), color='black')
plt.show()
```

```python
plt.figure(figsize=(3, 3)) #, dpi=300)
plt.grid(True)
plt.xticks([0, 1])
plt.yticks([0])
plt.plot(xs, -xs**3 * (1 - xs), color='black')
plt.show()
```
