```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

from math import sqrt

import sys
sys.path.append("../../scripts")

from model import Model, model_from_file
from model_plot import model_plot
from norms import *

%load_ext autoreload
%autoreload 2
%matplotlib inline
```

```python
data_dir = 'data_electric/'
```

Запустим один и тот же эксперимент с различной величиной заряда $q$. Посмотрим на падение напряжения и на профиль канала.

```python
model_moderate = model_from_file(data_dir + 'model_moderate.csv')
```

```python
model_plot(model_moderate, [0, 1e-5, 5e-5, 1e-3, 2e-2], figsize=(9, 6))
```

Естественное напряжение системы -- предел устанавливающегося напряжения при $q^+ \to +\infty$

```python
E_natural = 1.0 / model_moderate.params['l'] * sqrt(
    2.0 * model_moderate.params['Gamma'] * model_moderate.params['delta'] / model_moderate.params['eps_0']
)
E_natural
```

Заряд "перехода", при котором неповрежденное состояние становится неустойчивым

```python
q_transition = 2.0 * model_moderate.params['L'] / model_moderate.params['l'] * sqrt(
    model_moderate.params['eps_0'] * model_moderate.params['delta'] * model_moderate.params['Gamma']
)
q_transition
```

```python
print('{:e} {:e}'.format(model_moderate.E[0], model_moderate.E[-1]))
print('{:e}'.format(E_natural * sqrt(2.0)))
```

$E_N \cdot \sqrt{2}$ -- максимально возможное устанавливающееся напряжение 

```python
plt.figure(figsize=(9, 6))
plt.xlim([-1e-5, 1e-3])
plt.ylim([0, model_moderate.E[0] * 1.02])
plt.plot(model_moderate.ts, model_moderate.E)
plt.show()
```
