Пусть
$$\epsilon(\phi) = \cfrac{\epsilon_0}{g(\phi)},$$
$$g(\phi) = 4 \phi^3 - 3 \phi^4.$$
Рассмотрим $f(\phi) = 4 \phi^3 - 3 \phi^4$.

```python
model_pitike = model_from_file(data_dir + 'model_pitike.csv')
model_plot(model_pitike, [0, 1e-4, 9e-3, 1e-2, 2e-2], figsize=(9, 6))
```

```python
plt.figure(figsize=(9, 6))
plt.plot(model_pitike.ts, model_pitike.E)
plt.show()
```

```python
model_pitike.phi[-1].min()
```

```python
model_pitike.E[0] / model_pitike.E[-1]
```

```python
plt.figure(figsize=(10, 8))
plt.xlim([1.5e-5, 1.7e-5])
plt.ylim([-0.01, 0.6])
plt.plot(model_pitike.xs, model_pitike.phi[-1], '-o')
plt.show()
```

```python
def g(phi):
    return 4.0 * phi**3 - 3.0 * phi**4
```

```python
plt.figure(figsize=(9, 6))
plt.xlim([1.2e-5, 2.0e-5])
plt.plot(model_pitike.xs, g(model_pitike.phi[-1]))
plt.plot(model_pitike.xs, model_pitike.phi[-1])
plt.show()
```

```python
plt.figure(figsize=(9, 6))
plt.xlim([1.5e-5, 1.7e-5])
plt.yscale('log')
plt.plot(model_pitike.xs, 1.0 / g(model_pitike.phi[-1]), '-o')
plt.show()
```

Экспериментальная разработка: позволим $\phi \in (-\infty, 1]$ и положим
$$f(\phi) = \cfrac{1}{1 + (1 - \phi)^2},$$
$$g(\phi) = \cfrac{1}{1 + (1 - \phi)^2 + (1 - \phi)^p},$$
$p > 2$. Примем сейчас $p = 4$.

В расчете ниже напряжение нарочно взято на один порядок больше критического.

```python
def g_extended(phi):
    return 1.0 / (1.0 + (1.0 - phi)**2 + (1.0 - phi)**6)
```

```python
plt.figure(figsize=(9, 6))
plt.xlim([1.2e-5, 2.0e-5])
plt.plot(model_extended.xs, 1.0 / g_extended(model_extended.phi[-1]), '-o')
plt.show()
```
